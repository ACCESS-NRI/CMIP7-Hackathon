## ESMValTool Tiers

Observational datasets in ESMValTool are organised in tiers reflecting access restriction levels.

- Tier 1 Primarily Obs4MIPS where data is formatted, freely available and ready to use in ESMValTool.
- Tier 2 Data is freely available, CMORised datasets are available in ct11 in ESMValTool.
- Tier 3 These datasets have access rectrictions, licensing and acknowledgement may be required so direct access to the data cannot be provided. ACCESS-NRI can provide support to download and CMORise.

The `Dataset` object can be utilised through the esmvalcore API. To find all available datasets from ESGF which may not be available locally, set search_esgf to always. This example looks for all ensembles for a dataset.

API ref: https://docs.esmvaltool.org/projects/ESMValCore/en/latest/api/esmvalcore.dataset.html

[Discovering datasets example](https://docs.esmvaltool.org/projects/ESMValCore/en/latest/notebooks/discovering-data.html)

```python
# multiple observation comparisons

import iris

import iris.quickplot as qplt
import matplotlib.pyplot as plt

from esmvalcore.preprocessor import (
                regrid,
                annual_statistics,
                area_statistics )

iris.FUTURE.datum_support = True

def preproc_base(cube):
    # cube = extract_region(cube, 150, 270, -25, 25)
    cube = area_statistics(cube, 'mean')
    cube = annual_statistics(cube, 'mean')
    # cube = cube.collapsed('depth', iris.analysis.STD_DEV)
    print(cube.name, cube.units, cube.data.mean())
    return cube
    

preprocessed_cubes = {}
for obs_ds in list(obs.from_files()):
    # obs_ds['tier'] = obs_ds['tier'][-1]
    cube = preproc_base(obs_ds.load())
    iris.util.promote_aux_coord_to_dim_coord(cube, 'year')
    preprocessed_cubes[obs_ds['dataset']] = cube
    print(cube.shape)
```

```python
for lb, cube in preprocessed_cubes.items():
    # print(lb)
    qplt.plot(cube, label=lb)

plt.legend()
```
